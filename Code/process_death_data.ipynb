```python
import pandas as pd
import numpy as np
```

```python
death_data = pd.read_csv('deaths_NCHS_unprocessed.csv')
```

```python
death_data = death_data[["State", "MMWRYearWeek", "DeathsFromPneumoniaAndInfluenza"]]
```

```python
death_data[["MMWRYearWeek"]] = death_data[["MMWRYearWeek"]].astype(str)
```

Split into year and week

```python
death_data['Year'] = death_data['MMWRYearWeek'].str[:4].astype(int)
death_data['Week'] = death_data['MMWRYearWeek'].str[4:].astype(int)
death_data.drop(['MMWRYearWeek'], axis=1, inplace=True)
```

Add quarters

```python
quarter = [] 
for i in range(len(death_data)):
    if death_data['Week'][i] <= 13:
        quarter.append(1)
    elif death_data['Week'][i] <=26:
        quarter.append(2)
    elif death_data['Week'][i] <=39:
        quarter.append(3)
    elif death_data['Week'][i] <=53:
        quarter.append(4)
death_data["Quarter"] = quarter
```

Drop all rows where the state is NA

```python
death_data = death_data.dropna(subset=["State"]) 
```

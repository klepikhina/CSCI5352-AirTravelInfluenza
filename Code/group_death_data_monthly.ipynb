```python
import pandas as pd
```

```python
death_data = pd.read_csv('../Data/Clean/deaths_NCHS_processed.csv')
```

```python
death_data['YearWeek'] = death_data.Year.astype(str) + death_data.Week.astype(str)
death_data['YearWeek'] = death_data['YearWeek'].astype(int)
```

```python
death_data['LastDayWeek'] = pd.to_datetime((death_data['YearWeek']-1).astype(str) + "6", format="%Y%U%w")
```

```python
death_data['Month'] = pd.DatetimeIndex(death_data['LastDayWeek']).month
```

```python
death_data = death_data.drop(['YearWeek', 'LastDayWeek'], axis=1)
```
